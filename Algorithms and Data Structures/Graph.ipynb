```python
def BFS(graph, vertex):
    # Ready
    unvisited_vertices = [True]*len(graph[0])
    queue = []
    sequence = []

    # Start
    unvisited_vertices[vertex] = False
    queue.append(vertex)

    while len(queue) != 0:
        root = queue[0]
        queue.pop(0)
        sequence.append(root)

        for i in range(len(graph[vertex])):
            # Scan the adjacency vertex, the vertex must be unvisited.
            if i != root and unvisited_vertices[i] == True and graph[root][i] != 0:
                unvisited_vertices[i] = False
                queue.append(i)

    print(sequence)


BFS(G, 3)
```

```python
from collections import defaultdict
 
class Graph:

    def __init__(self, graph):

        self.graph = graph

    def DFSUtil(self, v, visited):
        visited.add(v)
        print(v, end=' ')
 
        for neighbour in range(len(self.graph[v])):
            if neighbour not in visited:
                self.DFSUtil(neighbour, visited)
 
    def DFS(self, v):
        visited = set()
        self.DFSUtil(v, visited)

    def display(self):
        print(self.graph)

graph = Graph(G)
graph.DFS(2)
# graph.display()
```
