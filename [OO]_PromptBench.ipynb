```python
import promptbench as pb
from tqdm import tqdm
```

```python
print('All supported datasets: ')
print(pb.SUPPORTED_DATASETS)
```

```python
dataset = pb.DatasetLoader.load_dataset("sst2")

dataset[:5]
```

```python
print('All supported models: ')
print(pb.SUPPORTED_MODELS)
```

```python
model = pb.LLMModel(model='google/flan-t5-large', max_new_tokens=10, temperature=0.001)
```

# Basic Usage

```python
# Prompt API supports a list, so you can pass multiple prompts at once.
role_based_prompt = ["As a sentiment classifier, determine whether the following text is 'positive' or 'negative'. Please classify: \nQuestion: {content}\nAnswer:"]
task_oriented_prompt = ["Classify the sentence as positive or negative: {content}",
                     "Determine the emotion of the following sentence as positive or negative: {content}"]
prompts = pb.Prompt(role_based_prompt)
```

```python
def proj_func(pred):
    mapping = {
        "positive": 1,
        "negative": 0
    }
    return mapping.get(pred, -1)
```

```python
for prompt in prompts:
    preds = []
    labels = []
    for data in tqdm(dataset):
        # process input
        input_text = pb.InputProcess.basic_format(prompt, data)
        label = data['label']
        raw_pred = model(input_text)

        # process output
        pred = pb.OutputProcess.cls(raw_pred, proj_func)
        preds.append(pred)
        labels.append(label)

    # evaluate
    score = pb.Eval.compute_cls_accuracy(preds, labels)
    print(f"{score:.3f}, {prompt}")
```
